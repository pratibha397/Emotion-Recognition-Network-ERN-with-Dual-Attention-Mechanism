```python
# Cell 1: Import necessary libraries
import os
import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
from PIL import Image
import torchvision.transforms as transforms
import torchvision.models as models
import matplotlib.pyplot as plt
import time
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import pandas as pd
import seaborn as sns
from tqdm import tqdm
import glob
import math
import random
import gc
from torch.cuda.amp import autocast, GradScaler

# Set seeds for reproducibility
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
torch.cuda.manual_seed(42)

# Enable PyTorch memory optimization
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Clear GPU memory
torch.cuda.empty_cache()
gc.collect()

# Cell 2: Define dataset paths for AffectNet YOLO format
dataset_path = '/kaggle/input/archive/YOLO_format'
train_dir = os.path.join(dataset_path, 'train')
val_dir = os.path.join(dataset_path, 'valid')
test_dir = os.path.join(dataset_path, 'test')

# Verify dataset structure
print(f"Train directory exists: {os.path.exists(train_dir)}")
print(f"Validation directory exists: {os.path.exists(val_dir)}")
print(f"Test directory exists: {os.path.exists(test_dir)}")

# Cell 3: Define constants and hyperparameters - ENHANCED FOR HIGH ACCURACY
BATCH_SIZE = 16  # Increased for better gradient estimation
ACCUMULATION_STEPS = 2  # Simulate batch size of 32 (16 * 2)
IMG_SIZE = 256  # Increased input size for better feature extraction
NUM_CLASSES = 8
DROPOUT = 0.4  # Slightly reduced for better learning
LEARNING_RATE = 0.00005  # Lower learning rate for more stable training
NUM_EPOCHS = 50  # More epochs for better convergence
WEIGHT_DECAY = 1e-4
PATIENCE = 15  # Increased patience to allow for more training
LABEL_SMOOTHING = 0.1  # For better generalization

# Emotion classes for AffectNet
EMOTIONS = ['neutral', 'happy', 'sad', 'surprise', 'fear', 'disgust', 'anger', 'contempt']

# Cell 4: Create dataset class for AffectNet YOLO format - ENHANCED
class AffectNetDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        
        # Check if the directory has 'images' and 'labels' subdirectories
        self.image_dir = os.path.join(root_dir, 'images')
        self.label_dir = os.path.join(root_dir, 'labels')
        
        # If 'images' and 'labels' subdirectories don't exist, assume the root_dir contains both
        if not os.path.exists(self.image_dir) or not os.path.exists(self.label_dir):
            self.image_dir = root_dir
            self.label_dir = root_dir
        
        # Get all image files
        image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
        self.image_files = []
        
        for ext in image_extensions:
            self.image_files.extend(glob.glob(os.path.join(self.image_dir, f'*{ext}')))
            self.image_files.extend(glob.glob(os.path.join(self.image_dir, f'*{ext.upper()}')))
        
        # Create a mapping from image filename to label
        self.label_map = {}
        
        # Get all label files
        label_extensions = ['.txt']
        label_files = []
        for ext in label_extensions:
            label_files.extend(glob.glob(os.path.join(self.label_dir, f'*{ext}')))
            label_files.extend(glob.glob(os.path.join(self.label_dir, f'*{ext.upper()}')))
        
        for label_file in label_files:
            # Extract base filename without extension
            base_name = os.path.splitext(os.path.basename(label_file))[0]
            
            # Find corresponding image file
            img_file = None
            for ext in image_extensions:
                possible_img = os.path.join(self.image_dir, f"{base_name}{ext}")
                if os.path.exists(possible_img):
                    img_file = possible_img
                    break
                possible_img_upper = os.path.join(self.image_dir, f"{base_name}{ext.upper()}")
                if os.path.exists(possible_img_upper):
                    img_file = possible_img_upper
                    break
            
            if img_file:
                # Read label from file (YOLO format: class_index x_center y_center width height)
                try:
                    with open(label_file, 'r') as f:
                        label_line = f.readline().strip()
                        if label_line:
                            label = int(label_line.split()[0])
                            self.label_map[img_file] = label
                except Exception as e:
                    print(f"Error reading label file {label_file}: {e}")
        
        # Create list of (image_path, label) pairs
        self.samples = []
        for img_path in self.image_files:
            if img_path in self.label_map:
                self.samples.append((img_path, self.label_map[img_path]))
        
        print(f"Loaded {len(self.samples)} samples from {root_dir}")
    
    def __len__(self):
        return len(self.samples)
    
    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        
        # Load image
        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            # Return a blank image if there's an error
            image = Image.new('RGB', (IMG_SIZE, IMG_SIZE), (0, 0, 0))
        
        # Apply transformations
        if self.transform:
            image = self.transform(image)
        
        return image, label

# Cell 5: Define advanced data transformations - ENHANCED and FIXED
# Custom RandAugment implementation with corrected functions
class RandAugment:
    def __init__(self, n=3, m=7):
        self.n = n
        self.m = m
        self.augment_list = [
            (self.autocontrast, 0, 1),
            (self.equalize, 0, 1),
            (self.rotate, -30, 30),
            (self.posterize, 0, 4),
            (self.solarize, 0, 256),
            (self.contrast, 0.1, 1.9),
            (self.brightness, 0.1, 1.9),
            (self.shear_x, -0.3, 0.3),
            (self.shear_y, -0.3, 0.3),
            (self.translate_x, -0.3, 0.3),
            (self.translate_y, -0.3, 0.3),
        ]
    
    def autocontrast(self, img, _):
        return transforms.functional.autocontrast(img)
    
    def equalize(self, img, _):
        return transforms.functional.equalize(img)
    
    def rotate(self, img, v):
        return transforms.functional.rotate(img, v)
    
    def posterize(self, img, v):
        v = int(v)
        return transforms.functional.posterize(img, v)
    
    def solarize(self, img, v):
        return transforms.functional.solarize(img, v)
    
    def contrast(self, img, v):
        return transforms.functional.adjust_contrast(img, v)
    
    def brightness(self, img, v):
        return transforms.functional.adjust_brightness(img, v)
    
    def shear_x(self, img, v):
        return transforms.functional.affine(img, 0, (0, 0), 1, (v, 0))
    
    def shear_y(self, img, v):
        return transforms.functional.affine(img, 0, (0, 0), 1, (0, v))
    
    def translate_x(self, img, v):
        return transforms.functional.affine(img, 0, (v * img.size[0], 0), 1, (0, 0))
    
    def translate_y(self, img, v):
        return transforms.functional.affine(img, 0, (0, v * img.size[1]), 1, (0, 0))
    
    def __call__(self, img):
        ops = random.choices(self.augment_list, k=self.n)
        for op, minval, maxval in ops:
            val = (float(self.m) / 30) * float(maxval - minval) + minval
            img = op(img, val)
        return img

train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE + 30, IMG_SIZE + 30)),
    transforms.RandomCrop(IMG_SIZE),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.15),
    transforms.RandomAffine(degrees=0, translate=(0.15, 0.15), scale=(0.85, 1.15)),
    RandAugment(n=3, m=7),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.4, scale=(0.02, 0.2))
])

test_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Cell 6: Create datasets and dataloaders - ENHANCED
train_dataset = AffectNetDataset(train_dir, transform=train_transform)
val_dataset = AffectNetDataset(val_dir, transform=test_transform)
test_dataset = AffectNetDataset(test_dir, transform=test_transform)

# DIAGNOSTIC CHECK: Check for overlap between train and validation sets
print("\n=== DIAGNOSTIC CHECKS ===")
print("Checking for overlap between train and validation sets...")
train_paths = set([sample[0] for sample in train_dataset.samples])
val_paths = set([sample[0] for sample in val_dataset.samples])
overlap = train_paths.intersection(val_paths)
print(f"Number of overlapping samples: {len(overlap)}")
if len(overlap) > 0:
    print("WARNING: There is overlap between training and validation sets!")
else:
    print("No overlap found between training and validation sets.")

# Check label distribution
train_labels = [sample[1] for sample in train_dataset.samples]
val_labels = [sample[1] for sample in val_dataset.samples]
test_labels = [sample[1] for sample in test_dataset.samples]

print("\nLabel distribution:")
print(f"Train: {np.bincount(train_labels)}")
print(f"Validation: {np.bincount(val_labels)}")
print(f"Test: {np.bincount(test_labels)}")

# Create dataloaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True)

print(f"\nTraining samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")
print(f"Testing samples: {len(test_dataset)}")

# Cell 7: Define the High-Accuracy Emotion Recognition Network (HAERN) with Novel Components
class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()
        self.conv = nn.Conv2d(2, 1, kernel_size, padding=kernel_size//2, bias=False)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        concat = torch.cat([avg_out, max_out], dim=1)
        out = self.conv(concat)
        return self.sigmoid(out)

class ChannelAttention(nn.Module):
    def __init__(self, in_channels, reduction_ratio=16):
        super(ChannelAttention, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)
        
        self.fc = nn.Sequential(
            nn.Linear(in_channels, in_channels // reduction_ratio),
            nn.ReLU(inplace=True),
            nn.Linear(in_channels // reduction_ratio, in_channels)
        )
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        b, c, _, _ = x.size()
        
        # Average path
        avg_out = self.fc(self.avg_pool(x).view(b, c))
        # Max path
        max_out = self.fc(self.max_pool(x).view(b, c))
        
        out = avg_out + max_out
        return self.sigmoid(out).view(b, c, 1, 1)

class CBAM(nn.Module):
    def __init__(self, in_channels, reduction_ratio=16, kernel_size=7):
        super(CBAM, self).__init__()
        self.channel_attention = ChannelAttention(in_channels, reduction_ratio)
        self.spatial_attention = SpatialAttention(kernel_size)
    
    def forward(self, x):
        out = x * self.channel_attention(x)
        out = out * self.spatial_attention(out)
        return out

class MultiScaleFeatureExtractor(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(MultiScaleFeatureExtractor, self).__init__()
        self.branch1 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )
        
        self.branch2 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )
        
        self.branch3 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=5, padding=2, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )
        
        self.branch4 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=7, padding=3, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )
        
        self.fusion = nn.Conv2d(out_channels * 4, out_channels, kernel_size=1, bias=False)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        
    def forward(self, x):
        b1 = self.branch1(x)
        b2 = self.branch2(x)
        b3 = self.branch3(x)
        b4 = self.branch4(x)
        
        out = torch.cat([b1, b2, b3, b4], dim=1)
        out = self.fusion(out)
        out = self.bn(out)
        out = self.relu(out)
        
        return out

class EmotionSpecificAttention(nn.Module):
    def __init__(self, in_channels, num_classes):
        super(EmotionSpecificAttention, self).__init__()
        self.num_classes = num_classes
        
        # Create attention masks for each emotion
        self.attention_masks = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(in_channels, in_channels // 4, kernel_size=1),
                nn.ReLU(inplace=True),
                nn.Conv2d(in_channels // 4, 1, kernel_size=1),
                nn.Sigmoid()
            ) for _ in range(num_classes)
        ])
        
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        
    def forward(self, x):
        batch_size, channels, height, width = x.size()
        
        # Generate attention maps for each emotion
        attention_maps = []
        for i in range(self.num_classes):
            attention_map = self.attention_masks[i](x)
            attention_maps.append(attention_map)
        
        # Stack attention maps
        attention_maps = torch.cat(attention_maps, dim=1)  # [B, num_classes, H, W]
        
        # Apply attention to features
        attended_features = []
        for i in range(self.num_classes):
            mask = attention_maps[:, i:i+1, :, :]
            attended = x * mask
            attended_features.append(attended)
        
        # Global pooling for each attended feature
        pooled_features = []
        for feature in attended_features:
            pooled = self.global_pool(feature)
            pooled_features.append(pooled.view(batch_size, -1))
        
        # Stack pooled features
        emotion_features = torch.stack(pooled_features, dim=1)  # [B, num_classes, channels]
        
        return emotion_features, attention_maps

class HAERN(nn.Module):
    """
    High-Accuracy Emotion Recognition Network (HAERN)
    Novel architecture with multi-scale feature extraction and emotion-specific attention
    """
    def __init__(self, num_classes=8):
        super(HAERN, self).__init__()
        
        # Backbone feature extractor (EfficientNet-B3)
        self.backbone = models.efficientnet_b3(weights=models.EfficientNet_B3_Weights.IMAGENET1K_V1).features
        
        # Multi-scale feature extraction
        self.multi_scale = MultiScaleFeatureExtractor(1536, 512)
        
        # CBAM attention
        self.cbam = CBAM(512)
        
        # Emotion-specific attention
        self.emotion_attention = EmotionSpecificAttention(512, num_classes)
        
        # Global feature aggregation
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        
        # Dropout for regularization
        self.dropout = nn.Dropout(0.5)
        
        # Final classifier
        self.classifier = nn.Sequential(
            nn.Linear(512, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.4),
            nn.Linear(256, 128),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )
        
    def forward(self, x, labels=None):
        # Extract backbone features
        x = self.backbone(x)
        
        # Multi-scale feature extraction
        x = self.multi_scale(x)
        
        # Apply CBAM attention
        x = self.cbam(x)
        
        # Emotion-specific attention
        emotion_features, attention_maps = self.emotion_attention(x)
        
        # Global pooling
        global_features = self.global_pool(x)
        global_features = global_features.view(global_features.size(0), -1)
        
        # Apply dropout
        global_features = self.dropout(global_features)
        
        # Classification
        logits = self.classifier(global_features)
        
        return logits, emotion_features, attention_maps, global_features, x

# Cell 8: Create the enhanced model
model = HAERN(num_classes=NUM_CLASSES)
model = model.to(device)

print(f"\nModel created with {sum(p.numel() for p in model.parameters())} parameters")
print(f"Trainable parameters: {sum(p.numel() for p in model.parameters() if p.requires_grad)}")

# Cell 9: Define enhanced loss function and optimizer - FIXED
class LabelSmoothingCrossEntropy(nn.Module):
    def __init__(self, num_classes, smoothing=0.1):
        super(LabelSmoothingCrossEntropy, self).__init__()
        self.num_classes = num_classes
        self.smoothing = smoothing
        self.confidence = 1.0 - smoothing
        
    def forward(self, pred, target):
        pred = pred.log_softmax(dim=-1)
        with torch.no_grad():
            true_dist = torch.zeros_like(pred)
            true_dist.fill_(self.smoothing / (self.num_classes - 1))
            true_dist.scatter_(1, target.data.unsqueeze(1), self.confidence)
        return torch.mean(torch.sum(-true_dist * pred, dim=-1))

class EmotionRecognitionLoss(nn.Module):
    def __init__(self, num_classes, alpha=0.3, smoothing=0.1):
        super(EmotionRecognitionLoss, self).__init__()
        self.num_classes = num_classes
        self.alpha = alpha
        self.smoothing = smoothing
        
        # Main classification loss with label smoothing
        self.cls_loss = LabelSmoothingCrossEntropy(num_classes, smoothing)
        
        # Projection layer for emotion features
        self.projection = nn.Linear(512, 1)
        
    def forward(self, outputs, targets):
        logits, emotion_features, _, _, _ = outputs
        
        # Main classification loss
        cls_loss = self.cls_loss(logits, targets)
        
        # Emotion-specific feature loss
        batch_size = emotion_features.size(0)
        emotion_features_flat = emotion_features.view(batch_size * self.num_classes, -1)  # [batch_size*num_classes, 512]
        
        # Project to scalar
        emotion_logits = self.projection(emotion_features_flat).squeeze(-1)  # [batch_size*num_classes]
        
        # Create binary labels for emotion-specific features (1 for correct emotion, 0 for others)
        binary_targets = torch.zeros(batch_size * self.num_classes, device=emotion_logits.device)
        for i in range(batch_size):
            # For each sample, set the correct emotion index to 1
            binary_targets[i * self.num_classes + targets[i]] = 1.0
        
        # Compute binary cross entropy loss
        emotion_loss = F.binary_cross_entropy_with_logits(emotion_logits, binary_targets)
        
        total_loss = cls_loss + self.alpha * emotion_loss
        
        return total_loss, cls_loss, emotion_loss

criterion = EmotionRecognitionLoss(NUM_CLASSES, alpha=0.3, smoothing=LABEL_SMOOTHING)
criterion = criterion.to(device)

# Use AdamW optimizer with weight decay
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

# Use cosine annealing with warm restarts
scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=2, eta_min=1e-6)

# Cell 10: Train the enhanced model with gradient accumulation and mixed precision
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=50):
    # Fixed GradScaler initialization - using new API
    scaler = torch.amp.GradScaler('cuda')
    train_loss_history = []
    train_acc_history = []
    val_loss_history = []
    val_acc_history = []
    
    best_val_acc = 0.0
    best_model_wts = model.state_dict()
    patience_counter = 0
    
    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)
        
        # Training phase
        model.train()
        running_loss = 0.0
        running_corrects = 0
        
        optimizer.zero_grad()
        
        train_pbar = tqdm(train_loader, desc="Training", unit="batch", leave=True)
        for batch_idx, (inputs, labels) in enumerate(train_pbar):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            # Fixed autocast usage - using new API
            with torch.amp.autocast('cuda'):
                outputs = model(inputs, labels)
                loss, _, _ = criterion(outputs, labels)
                loss = loss / ACCUMULATION_STEPS
            
            # Backward pass with gradient scaling
            scaler.scale(loss).backward()
            
            # Update weights every ACCUMULATION_STEPS
            if (batch_idx + 1) % ACCUMULATION_STEPS == 0:
                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad()
            
            # Statistics
            running_loss += (loss.item() * ACCUMULATION_STEPS) * inputs.size(0)
            
            _, preds = torch.max(outputs[0], 1)
            running_corrects += torch.sum(preds == labels.data)
            
            # Update progress bar
            train_pbar.set_postfix({
                'loss': loss.item() * ACCUMULATION_STEPS,
                'acc': (torch.sum(preds == labels.data).double() / inputs.size(0)).item()
            })
            
            # Clear cache periodically
            if batch_idx % 50 == 0:
                torch.cuda.empty_cache()
        
        # Update for any remaining gradients
        scaler.step(optimizer)
        scaler.update()
        optimizer.zero_grad()
        
        train_pbar.close()
        
        epoch_loss = running_loss / len(train_loader.dataset)
        epoch_acc = running_corrects.double() / len(train_loader.dataset)
        
        train_loss_history.append(epoch_loss)
        train_acc_history.append(epoch_acc.item())
        
        print(f'Train Loss: {epoch_loss:.4f}')
        print(f'Train Acc: {epoch_acc:.4f}')
        
        # Validation phase with diagnostic checks
        model.eval()
        running_loss = 0.0
        running_corrects = 0
        
        val_pbar = tqdm(val_loader, desc="Validation", unit="batch", leave=True)
        with torch.no_grad():
            for inputs, labels in val_pbar:
                inputs = inputs.to(device)
                labels = labels.to(device)
                
                # Fixed autocast usage - using new API
                with torch.amp.autocast('cuda'):
                    outputs = model(inputs, labels)
                    loss, _, _ = criterion(outputs, labels)
                
                running_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs[0], 1)
                running_corrects += torch.sum(preds == labels.data)
                
                val_pbar.set_postfix({
                    'loss': loss.item(),
                    'acc': (torch.sum(preds == labels.data).double() / inputs.size(0)).item()
                })
        
        val_pbar.close()
        torch.cuda.empty_cache()
        
        epoch_loss = running_loss / len(val_loader.dataset)
        epoch_acc = running_corrects.double() / len(val_loader.dataset)
        
        val_loss_history.append(epoch_loss)
        val_acc_history.append(epoch_acc.item())
        
        print(f'Val Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
        
        # Diagnostic check for suspicious accuracy
        if epoch_acc == 1.0:
            print("WARNING: 100% validation accuracy detected! This might indicate data leakage or overfitting.")
        
        # Update learning rate
        scheduler.step()
        
        # Save the best model
        if epoch_acc > best_val_acc:
            best_val_acc = epoch_acc
            best_model_wts = model.state_dict()
            torch.save(model.state_dict(), 'best_emotion_model.pth')
            print(f"New best model saved with accuracy: {best_val_acc:.4f}")
            patience_counter = 0
        else:
            patience_counter += 1
        
        # Early stopping
        if patience_counter >= PATIENCE:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break
        
        print()
    
    # Load best model weights
    model.load_state_dict(best_model_wts)
    return model, train_loss_history, train_acc_history, val_loss_history, val_acc_history

# Cell 11: Train the enhanced model
print("\nStarting training with enhanced model...")
model, train_loss, train_acc, val_loss, val_acc = train_model(
    model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=NUM_EPOCHS
)

# Cell 12: Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Loss History')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Accuracy History')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.savefig('/kaggle/working/training_history.png')
plt.show()

# Cell 13: Evaluate the enhanced model on test set
def evaluate_model(model, test_loader):
    model.eval()
    all_preds = []
    all_labels = []
    
    eval_pbar = tqdm(test_loader, desc="Evaluating", unit="batch", leave=True)
    with torch.no_grad():
        for inputs, labels in eval_pbar:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            # Fixed autocast usage - using new API
            with torch.amp.autocast('cuda'):
                outputs = model(inputs, labels)
                _, preds = torch.max(outputs[0], 1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    eval_pbar.close()
    
    return all_preds, all_labels

print("\nEvaluating enhanced model on test set...")
all_preds, all_labels = evaluate_model(model, test_loader)

# Calculate accuracy
test_accuracy = accuracy_score(all_labels, all_preds)
print(f"Test Accuracy: {test_accuracy:.4f}")

# Compute confusion matrix
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=EMOTIONS, yticklabels=EMOTIONS)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.savefig('/kaggle/working/confusion_matrix.png')
plt.show()

# Print classification report
print("\nClassification Report:")
print(classification_report(all_labels, all_preds, target_names=EMOTIONS))

# Cell 14: Save final model and results
torch.save(model.state_dict(), '/kaggle/working/final_emotion_model.pth')

# Save evaluation results
results = {
    'test_accuracy': test_accuracy,
    'confusion_matrix': cm.tolist(),
    'classification_report': classification_report(all_labels, all_preds, target_names=EMOTIONS, output_dict=True)
}

import json
with open('/kaggle/working/evaluation_results.json', 'w') as f:
    json.dump(results, f)

print("Model and results saved successfully.")

# Cell 15: Visualize feature maps and attention
def visualize_features_and_attention(model, test_loader, num_samples=3):
    model.eval()
    samples_done = 0
    
    plt.figure(figsize=(20, 15))
    
    with torch.no_grad():
        for inputs, labels in test_loader:
            if samples_done >= num_samples:
                break
                
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            # Get model outputs
            outputs = model(inputs[:1], labels[:1])
            _, _, attention_maps, global_features, backbone_features = outputs
            
            # Process first sample in batch
            img = inputs[0].cpu().numpy().transpose(1, 2, 0)
            img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
            img = np.clip(img, 0, 1)
            
            # Plot original image
            plt.subplot(num_samples, 5, samples_done*5 + 1)
            plt.imshow(img)
            plt.title(f"True: {EMOTIONS[labels[0]]}")
            plt.axis('off')
            
            # Plot backbone features (first channel)
            plt.subplot(num_samples, 5, samples_done*5 + 2)
            plt.imshow(backbone_features[0, 0].cpu().numpy(), cmap='viridis')
            plt.title("Backbone Features")
            plt.axis('off')
            
            # Plot attention map for predicted emotion
            _, pred = torch.max(outputs[0], 1)
            pred_emotion = pred[0].item()
            plt.subplot(num_samples, 5, samples_done*5 + 3)
            plt.imshow(attention_maps[0, pred_emotion].cpu().numpy(), cmap='viridis')
            plt.title(f"Attention: {EMOTIONS[pred_emotion]}")
            plt.axis('off')
            
            # Plot global features
            plt.subplot(num_samples, 5, samples_done*5 + 4)
            global_feature = global_features[0].cpu().numpy().reshape(1, -1)
            plt.imshow(global_feature, cmap='viridis', aspect='auto')
            plt.title("Global Features")
            plt.axis('off')
            
            # Plot prediction probabilities
            plt.subplot(num_samples, 5, samples_done*5 + 5)
            probs = F.softmax(outputs[0], dim=1)[0].cpu().numpy()
            plt.bar(range(NUM_CLASSES), probs)
            plt.xticks(range(NUM_CLASSES), EMOTIONS, rotation=45)
            plt.title("Prediction Probabilities")
            plt.tight_layout()
            
            samples_done += 1
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/features_and_attention.png')
    plt.show()

print("\nVisualizing features and attention maps...")
visualize_features_and_attention(model, test_loader)

# Cell 16: Feature space visualization
from sklearn.manifold import TSNE

def visualize_feature_space(model, test_loader, num_samples=500):
    model.eval()
    features = []
    labels = []
    
    with torch.no_grad():
        for inputs, targets in test_loader:
            if len(features) >= num_samples:
                break
                
            inputs = inputs.to(device)
            
            # Fixed autocast usage - using new API
            with torch.amp.autocast('cuda'):
                outputs = model(inputs, targets)
                _, _, _, global_features, _ = outputs
            
            features.append(global_features.cpu().numpy())
            labels.append(targets.numpy())
    
    features = np.concatenate(features, axis=0)[:num_samples]
    labels = np.concatenate(labels, axis=0)[:num_samples]
    
    # Apply t-SNE
    tsne = TSNE(n_components=2, random_state=42)
    embedded_features = tsne.fit_transform(features)
    
    # Plot
    plt.figure(figsize=(12, 10))
    for i, emotion in enumerate(EMOTIONS):
        idx = labels == i
        plt.scatter(embedded_features[idx, 0], embedded_features[idx, 1], label=emotion, alpha=0.7, s=50)
    
    plt.title("Feature Space Visualization (t-SNE)")
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.legend()
    plt.savefig('/kaggle/working/feature_space.png')
    plt.show()

print("\nVisualizing feature space...")
visualize_feature_space(model, test_loader)

# Cell 17: Generate final report
print("\n" + "="*50)
print("FINAL MODEL EVALUATION REPORT")
print("="*50)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Total Model Parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"Trainable Parameters: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")
print("\nBest Model saved to: /kaggle/working/best_emotion_model.pth")
print("Final Model saved to: /kaggle/working/final_emotion_model.pth")
print("Evaluation results saved to: /kaggle/working/evaluation_results.json")
print("Training history plot saved to: /kaggle/working/training_history.png")
print("Confusion matrix saved to: /kaggle/working/confusion_matrix.png")
print("Features and attention visualization saved to: /kaggle/working/features_and_attention.png")
print("Feature space visualization saved to: /kaggle/working/feature_space.png")
print("="*50)
```
